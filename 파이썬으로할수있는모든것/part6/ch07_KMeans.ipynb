# KMeans - 고객 분류하기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 연습용 데이터 만들기

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, centers=3, random_state=100)
X
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
y
```

```python
sns.scatterplot(data=X, x=0, y=1, hue=y, palette='Set2')
```

## 연습용 데이터에 Kmeans Clustering 활용하기

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(X)
```

```python
pred = km.predict(X)
```

```python
sns.scatterplot(data=X, x=0, y=1, hue=pred, palette='Set2')
plt.show()
```

```python
km = KMeans(n_clusters=4)
km.fit(X)
pred = km.predict(X)
```

```python
sns.scatterplot(data=X, x=0, y=1, hue=pred, palette='Set2')
plt.show()
```

## Elbow Method로 최적의 Clustering 객수 찾기

```python
km.inertia_
```

```python
inertia_list = []
for n in range(2,11):
    km = KMeans(n_clusters=n)
    km.fit(X)
    inertia_list.append(km.inertia_)
inertia_list
```

```python
sns.lineplot(x=range(2,11), y=inertia_list)
plt.show()
```

## 데이터 살펴보기

```python
url = 'https://github.com/DSNote/fastcampus/raw/main/marketing.csv'
```

```python
mkt_df = pd.read_csv(url)
mkt_df.head()
```

```python
mkt_df = mkt_df.drop('ID', axis=1)
mkt_df.info()
```

```python
mkt_df.describe()
```

```python
mkt_df = mkt_df[mkt_df['Year_Birth']>=1900]
```

```python
mkt_df['Income'].sort_values(ascending=False)
```

```python
mkt_df = mkt_df[mkt_df['Income']!=666666.0]
mkt_df.describe()
```

## 결속치 처리

```python
mkt_df.isna().mean()
```

```python
mkt_df = mkt_df.dropna()
```

## 시간 데이터 변환

```python
mkt_df['Dt_Customer'] = pd.to_datetime(mkt_df['Dt_Customer'])
```

```python
mkt_df['Dt_Customer'].max()
```

```python
mkt_df['cust_age'] = mkt_df['Dt_Customer'].max().year*12+mkt_df['Dt_Customer'].max().month - (mkt_df['Dt_Customer'].dt.year*12+mkt_df['Dt_Customer'].dt.month)
mkt_df = mkt_df.drop('Dt_Customer', axis=1)
```

## 피쳐 엔지니어링

```python
mkt_df['Total_mnt'] =  mkt_df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
```

```python
mkt_df['Children'] =  mkt_df[['Kidhome', 'Teenhome']].sum(axis=1)
```

```python
mkt_df['Marital_Status'].value_counts()
```

```python
mkt_df['Marital_Status'] = mkt_df['Marital_Status'].replace({'Married':'Partner', 'Together':'Partner', 'Single':'Single', 'Divorced':'Single', 'Widow':'Single', 'Alone':'Single', 'Absurd':'Single', 'YOLO':'Single'})
```

```python
mkt_df['Marital_Status'].value_counts()
```

```python
mkt_df['Education'].value_counts()
```

```python
mkt_df['Education'] = mkt_df['Education'].replace({'Basic':'Undergraduate','2n Cycle':'Undergraduate','Graduation':'Graduate', 'Master':'Postgraduate', 'PhD':'Postgraduate'})
```

```python
mkt_df['Education'].value_counts()
```

```python
mkt_df = mkt_df.drop(['Kidhome', 'Teenhome'], axis=1)
```

## 더미 변수 만들기

```python
mkt_df = pd.get_dummies(data=mkt_df, columns=['Education', 'Marital_Status'])
```

```python
mkt_df.head()
```

## Kmeans 알고리즘으로 군집 나누기

```python
inertia_list = []
for n in range(2,11):
    km = KMeans(n_clusters=n, random_state=100)
    km.fit(mm_df)
    inertia_list.append(km.inertia_)
```

```python
inertia_list
```

```python
sns.lineplot(x=range(2,11), y=inertia_list)
plt.show()
```

## 실루엣 스코어로 최적의 Clustering 갯수 찾기

```python
from sklearn.metrics import silhouette_score
```

```python
score = []
for n in range(2,11):
    km = KMeans(n_clusters=n, random_state=100)
    km.fit(mm_df)
    pred = km.predict(mm_df)
    score.append(silhouette_score(mm_df, pred))
```

```python
sns.lineplot(x=range(2,11), y=score)
plt.show()
```

```python
km = KMeans(n_clusters=5, random_state=100)
```

```python
km.fit(mm_df)
```

```python
pred = km.predict(mm_df)
```
