```python
# 막대 그래프
plt.figure(figsize=(8,4))
# plt.grid(alpha=0.3, color='black', linewidth=0.5)
plt.title('Coffee Sales in March', fontsize=20, loc='center')
plt.xlabel('Menu', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.bar(x, y, color='lightcoral', label='Sales')
plt.plot([100,200,400,800,1600][::-1], label='Average')
plt.legend(loc='best')
plt.ylim(0,2000)
plt.yticks([num for num in range(0, 2000, 200)])
plt.show()
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
# Histplot
sns.histplot(data=data, x='body_mass_g', bins=30)
```

```python
sns.histplot(data=data, x='body_mass_g', bins=30, hue='species')
```

```python
sns.histplot(data=data, x='body_mass_g', bins=30, hue='species', multiple='stack')
```

```python
plt.figure(figsize=(6,6))
plt.title('Distribution over body_mass', fontsize=14)
sns.histplot(data=data, x='body_mass_g', bins=30, hue='species')
plt.show()
```

```python
# Displot
# kind = hist, kde, ecdf
sns.displot(data=data, kind='kde', x='body_mass_g', hue='species')
```

```python
sns.displot(data=data, kind='kde', x='body_mass_g', hue='species', col='island')
```

```python
sns.displot(data=data, kind='kde', x='body_mass_g', hue='species', row='island')
```

```python
# Barplot
sns.barplot(data=data, x='species', y='body_mass_g')
plt.show()
```

```python
sns.barplot(data=data, x='body_mass_g', y='species')
plt.show()
```

```python
sns.barplot(data=data, x='species', y='body_mass_g', hue='sex')
plt.show()
```

```python
# Countplot
sns.countplot(data=data, x='sex')
plt.show()
```

```python
sns.countplot(data=data, x='sex', hue='species')
plt.show()
```

```python
sns.set_palette('viridis')
sns.countplot(data=data, x='sex', hue='species')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8,6))
sns.set_palette('Set2')
sns.boxplot(data=data, x='species', y='bill_depth_mm')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(6,4))
sns.set_palette('Set2')
sns.boxplot(data=data, y='bill_depth_mm')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_palette('Set2')
sns.boxplot(data=data, x='species', y='bill_depth_mm', hue='sex')
plt.show()
```

```python
# Violinplot
sns.violinplot(data=data, x='species', y='bill_depth_mm', hue='sex')
plt.show()
```

```python
# Lineplot
sns.lineplot(data=data, x='body_mass_g', y='flipper_length_mm')
plt.show()
```

```python
# Pointplot
sns.pointplot(data=data, x='species', y='bill_length_mm')
plt.show()
```

```python
# Scatterplot
sns.scatterplot(data=data, x='flipper_length_mm', y='body_mass_g', hue='sex')
plt.show()
```

```python
# Pairplot
sns.pairplot(data=data, hue='species')
plt.show()
```

```python
sns.pairplot(data=data, hue='sex')
plt.show()
```

```python
# Heatmap
corr = data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].corr()
sns.heatmap(data=corr, square=True, cmap='Blues', annot=True, fmt='.3f')
plt.show()
```
